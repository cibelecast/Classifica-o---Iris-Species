```python
#Utilização do algoritmo de classificação chamado K-NearestNeighbors. Este algoritmo para fazer uma previsão
#para um novo "ponto de data", ele acha o ponto no conjunto de dados de treinamento que é mais proximo desse novo ponto. E então ele 
#o designa para esse rotulo de ponto de treinamento. O K significa que pode ser o K, (os 3 ou 5 pontos vizinhos dos dados de treinamento)
```

```python
#O objeto KNN encapsula o algoritmo que vai ser utilizado na construção do nosso modelo para o treinamento dos dados.
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1) # seta o número de neighbors = 1
```

```python
#Esta função train_test_split divide o conjunto de dados, em conjunto de dados de treino e conjunto de dados de teste.
#Divide-se o nosso dataset em conjunto de dados de treino para construir o modelo, e o outro conjunto de dados de teste
#é para avaliar se o modelo terá um bom desempenho com novos dados(*que não foram vistos antes).Desse modo, temos o  
#benefício de se ter dados de teste para avaliação do desempenho do modelo para evitar,por exemplo, o problema de 
#overfitting(erro de variância) dos dados. Se não houver esta divisão dos datasets, o seu modelo pode não ter precisão
#e passar a ser inadequado para a determinada solução, pois não seria capaz de prever novos dados, e por isso iria 
#falhar.
# O X deve ser entendido como os dados de entrada e Y deve ser entendido como os dados de saida. X_train y_train vão ser
#utilizados pelo classificador da função e X_teste e y_teste vão ser utilizados pelo parametro score para poder realizar
#a avaliação do modelo.
from sklearn.model_selection import train_test_split
X_train,X_teste,y_train,y_teste=train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=0)
    
#Neste caso utilizou o 'data', pois ele contém os dados das 4 colunas, e isso que vai ser treinando
#o target é o que você quer prever, é a sua saída. No caso este target, contem numeros de 0 a 2,cada numero representa
#uma classe de "iris" dentro desse modelo
#Esta função train_test_split utiliza o random_state para embaralhar os dados antes de dividi-los, senão, seria só dividir
#na metade, e ai isso não iria possibilitar do modelo realizar sua previsão, para mais detalhes leia o livro. Pois, não
#iria abranger todas as classes.
#como você não colocou o parametro:test_size=0.2, ele assume automaticamente que o tamanho do test_set será de 25 por cento
# e o tamanho do training_set será de 75 por cento.
# X_train contem 75 por cento dos dados
# X_Teste contém 25 por cento dos dados
```

```python
#Para construir o modelo do conjunto de treinamento, nos chamamos o metodo knn, 
#que tem como argumento o X-train(*que sao aqueles 4 dados. ELe contem os dados de treinamento) e o y_train
#(*que contem os training labels, que nada mais sao que os nomes das "iris".)

knn.fit(X_train,y_train)

#ele de da como output o objeto knn, agora nos temos uma representação do nosso classificador. Todos os parametros,
#exceto o n_neighbors=1, são os valores default e podem ser modificados de acordo com a necessidade.
```

```python
#Agora que nos já temos nosso modelo construido. Faremos previsões em NOVOS DADOS que nao sabemos 
#quais serao os labels(neste caso os nomes das "iris")
import numpy as np
X_new=np.array([[5,2.9,1,0.2]])
```

```python
X_new
```

```python
print("X_new.shape: {}").format(X_new.shape) #matrix 1 linha e 4 colunas, 1x4
```

```python
#Numpy sempre espera array com 2 dimensões para os dados
#Para fazer uma previsão, nos chamamos o metodo predict do objeto knn:
prediction = knn.predict(X_new)
print("Prediction: {}".format(prediction)) # neste caso X new, teve y_train = 0, recebeu o label 0
```

```python
print("Predicted target name: {}".format(iris_dataset['target_names'][prediction])) 
# relaciona o target_name com o 0, obtido anteriormente
```

```python
#Como vamos saber se podemos confiar nessa previsão? Por isso, devemos realizar uma AVALIAÇÂO DO MODELO< só assim
#poderemos confiar se o modelo previu corretamente.

#Nesta parte entra os datasets de TESTE, que nos criamos anteriormente.Estes dados não foram construidos para construir
#o modelo.
#Assim sendo nos vamos fazer uma previsão de quais seriam os resultados desse conjunto de dados, e vamos comparar esse resultado
#com o "verdadeiro resultado" que se encontra no y_test e assim teremos estatisticamente um valor que vai determinar
# se o modelo é confiavel ou não, geralmente o valor acima de 70 por cento já quer dizer que o modelo é confiavel.

y_pred=knn.predict(X_teste)
#impressão do resultado da previsão do X_teste
print("Test set predictions:\n {}".format(y_pred))
```

```python
print("Test set score: {:.2f}".format(np.mean(y_pred == y_teste)))
```

```python
#Nós também podemos usar um metodo chamado SCORE do objeto KNN
print("Test set score: {:.2f}".format(knn.score(X_teste,y_teste)))
```
